```python
import scipy
import scipy.fftpack
import numpy as np
import librosa
import librosa.display
from IPython.display import Audio
import matplotlib.pyplot as plt
```

```python
#
```

```python
# Create Pitch
def pitchMaker(frequency=262,magnitude = 1,time = 1,sr = 22050):
    t = np.arange(0, time,1/sr)
    data = magnitude*np.sin(2*np.pi*frequency*t)    
    return data
```

```python
def plotTimeDomain(data,sr=22050):
    max_points = 1000
    librosa.display.waveplot(data[:max_points], sr=44100)
```

```python
def plotFreqDomain(freq,sr=22050):
    freq_plot = np.abs(freq)[::10]
    plt.plot(freq_plot)
    num_ticks = 10
    ticks = len(freq_plot)/10*np.arange(10)
    plt.xticks(ticks, sr/len(freq)*ticks*10)
    plt.show()
def plotFreqDomain_zoom(freq,sr=22050):
    freq1 = freq[:1000]
    freq_plot = np.abs(freq1)[::10]
    plt.plot(freq_plot)
    num_ticks = 10
    ticks = len(freq_plot)/10*np.arange(10)
    plt.xticks(ticks, (sr/len(freq)*ticks*10))
    plt.show()
```

```python
data = pitchMaker()
```

```python
plotTimeDomain(data)
```

```python
Audio(data, rate=22050)
```

```python
freq = scipy.fftpack.fft(data)
```

```python
plotFreqDomain(freq)
```

```python
plotFreqDomain_zoom(freq)
```
